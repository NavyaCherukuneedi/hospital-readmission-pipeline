```python
import pandas as pd
import os
from pathlib import Path

# Get absolute path to data folder
data_path = Path('/Users/srujana/Desktop/HealthCare_Datapipeline/hospital-readmission-pipeline/data')

print(f"Looking for data in: {data_path}")
print(f"Path exists: {data_path.exists()}")

# List all files in data folder
all_files = os.listdir(data_path)
print(f"All files in data folder: {all_files}")

# Find CSV files
csv_files = [f for f in all_files if f.endswith('.csv')]
print(f"Found CSV files: {csv_files}")

if csv_files:
    # Load the main dataset
    df = pd.read_csv(data_path / csv_files[0])

    # Basic exploration
    print(f"\nDataset shape: {df.shape}")
    print(f"\nColumn names and types:")
    print(df.dtypes)
    print(f"\nFirst few rows:")
    print(df.head())
    print(f"\nMissing values:")
    print(df.isnull().sum())
else:
    print("No CSV files found!")
```
